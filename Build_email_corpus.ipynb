```python
with open('/content/drive/MyDrive/small_mail_step1.txt', 'r') as f:
  small_text = f.read()
```

```python
small_text = re.sub(r"[<>\?\"\':;()\$@\!\`\-_\+\=&\%\#\{\}]", "", small_text)
```

```python
tokenized = re.split(r"\s", small_text)
```

```python
tokenized[:24]
```

```python
small_tokens = []
for token in tokenized:
  if len(token) > 30:
    continue
  if len(re.findall(r"https", token)) > 0:
    continue
  if token == "":
    continue
  small_tokens.append(token)
small_text = '\n'.join(small_tokens)
with open('/content/drive/MyDrive/small_mail_step2.txt', 'w') as f:
  f.write(small_text)
```
